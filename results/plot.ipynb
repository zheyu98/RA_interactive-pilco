```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 1000)
c = np.random.choice(x, size=10, replace=False)
c2 = np.random.choice(x, size=10, replace=False)
y = np.sin(c) + np.random.normal(scale=0.05, size=len(c))
y2 = np.sin(c2) + 0.3 + np.random.normal(scale=0.05, size=len(c2))
```

```python
z = np.polyfit(c, y, 3)
f = np.poly1d(z)
yn = f(x)

cc = np.concatenate((c, c2))
yy = np.concatenate((y, y2))
z2 = np.polyfit(cc, yy, 20)
f2 = np.poly1d(z2)
yn2 = f2(x)
```

```python
plt.ylim(0.0, 1.1)
plt.plot(c, y, '.', markersize=12)
plt.plot(c2, y2, '.', markersize=12)
plt.plot(x, yn, '--')
plt.plot(x, yn2, '--')
plt.legend(["Model 1 data", "Model 2 data", "Model 1 fit", "Combined model fit"])
plt.xlabel(r"(x_t, u_t)")
plt.ylabel(r"$\Delta_t$")
plt.savefig("model_change.png")
#plt.()
```
